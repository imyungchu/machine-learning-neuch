```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = np.arange(0,1,0.01)
plt.plot(q, q / (10 - 9*q))
plt.xlabel("Prior probability of innoncence")
plt.ylabel("Posterior")
```

```python
def get_posterior(prior, data, likelihood):
    marginal = prior * likelihood[data][0] + (1 - prior) * likelihood[data][1]
    posterior = prior * likelihood[data][0] / marginal
    return posterior

prior = 0.9 # Pr(H0)
likelihood = np.zeros([2, 2])
likelihood[0][0] = 0.6 # Pr(F|H0)
likelihood[1][0] = 0.4 # Pr(T|H0)
likelihood[0][1] = 0.01 # Pr(F|H1)
likelihood[1][1] = 0.99 # Pr(T|H1)
data = 1
N=100
negative_posterior = np.zeros(N)
positive_posterior = np.zeros(N)
prior = np.arange(N)/N
for t in range(N):
    positive_posterior[t] = get_posterior(prior[t], 1, likelihood)
    negative_posterior[t] = get_posterior(prior[t], 0, likelihood)
```

```python
plt.plot(prior, positive_posterior)
plt.plot(prior, negative_posterior)
```
