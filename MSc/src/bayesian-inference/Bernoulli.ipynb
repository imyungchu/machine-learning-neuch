```python
theta = 0.6
import numpy as np
n_data = 10
data = np.random.choice(2, p = [1 - theta, theta], size=n_data)
```

```python
Thetas = np.arange(0,1,0.01)
def likelihood(theta, data):
    s_T = np.sum(data)
    T = len(data)
    return np.power(theta, s_T)*np.power((1 - theta), T - s_T)
```

```python
import matplotlib.pyplot as plt
plt.plot(Thetas, likelihood(Thetas, data))
plt.xlabel("$\\theta$")
plt.ylabel("$P(data | \\theta)$")
data
```

```python
# Calculate P(D|theta) P(theta) / \sum_{\theta'} P(D|theta') P(theta')
def get_posterior(prior, Thetas, probs, data):
    n_thetas = len(Thetas)
    posterior = prior.copy()
    for k in range(n_thetas):
        posterior[k] = prior[k] * probs(Thetas[k], data)
    posterior /= np.sum(posterior)
    return posterior
```

```python
n_data = 10
data = np.random.choice(2, p = [1 - theta, theta], size=n_data)
prior = np.exp(-10*(Thetas - 0.5)**2)  
prior /= sum(prior)
posterior = get_posterior(prior, Thetas, likelihood, data)
plt.plot(Thetas, prior, '-')
plt.plot(Thetas, posterior, '--')
plt.plot(Thetas, likelihood(Thetas, data), ':')
np.mean(data)
```
