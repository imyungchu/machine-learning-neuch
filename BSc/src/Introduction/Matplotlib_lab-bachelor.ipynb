## Line Chart


A line chart displays information as a series of data points or markers, Line charts work out of the box with matplotlib. You can have multiple lines in a line chart, change color, change type of line and much more

let's take an examlpe to understand and translate the idea of line chart. Let's consider we have a table representing the velocity of 2 moving object at different instance of time and we want to draw the velocity - time graph.

```python
## firt , create time list 
time = [0 ,5,10,15,20,25,30]

## create velocity of the first object 
velocity_obj_1 = [10,15,20,20,30,15,0]

## create velocity of the second object 
velocity_obj_2 = [0,12,15,12,20,10,25]
```

 Create velocity of the first by the time

```python
## we use plot function to drow line chart 
plt.plot(time  , velocity_obj_1) 
```

 Create velocity of the second by the time

```python
## we use plot function to drow line chart 
plt.plot(time , velocity_obj_2)
```

merge the both by using plt.plot 2 times

```python
## let's merge the 2 velocity of 2 objects by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)
```

[<matplotlib.lines.Line2D at 0x1aaeba93af0>] , we can remove this line by putting ; at the end of last line , or using show function.

```python
## let's merge the 2 velocity of 2 objects by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2);
#plt.show()
```

we can add title of our graph by using plt.title() , and customize the size of font by using fontsize attribute

```python
# we can put title  by using plt.title() , and customize the size of font by using fontsize attribute 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

plt.show()
```

Add legend

```python
## let's merge both by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1)
plt.plot(time  , velocity_obj_2)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

# add legend
plt.legend(["object 1 " , "object 2 "], loc='upper right') ;
```

Add x and y label and linewidth of each line

```python
## let's merge both by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1, linewidth = 3)
plt.plot(time  , velocity_obj_2, linewidth = 3)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

# add legend
plt.legend(["object 1 " , "object 2 "], loc='upper left') 

plt.xlabel("Time" , fontsize = 12 )
plt.ylabel("velocity of objects" , fontsize = 12 ) ;
```

To make the graph bigger, we use a function called figure 

```python
# graph size 
plt.figure(figsize=(15 , 10 )) # the first attribute is the width and and the second one is the length

## let's merge both by using plt.plot 2 times 
plt.plot(time  , velocity_obj_1, linewidth = 3)
plt.plot(time  , velocity_obj_2, linewidth = 3)

# Title
plt.title("velocity of 2 objects " , fontsize = 20 )

# add legend
plt.legend(["object 1 " , "object 2 "], loc='upper left') 

plt.xlabel("Time" , fontsize = 12 )
plt.ylabel("velocity of objects" , fontsize = 12 ) ;
```

## Scatter Plot

In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additonally, you can also use a third variable to determine the size or color of the points. let's do previous example with scatter plot

```python
# use plt.scatter to create a scatter graph
# c refers to the collor points 
# s refers to size of points  
# alpha refers to color opacity of each point , number of alpha must between 0.1 and 1  
plt.scatter(time , velocity_obj_1 , c= 'r' , s = 100 , alpha= 0.8)
plt.scatter(time , velocity_obj_2 , c= 'k' , s = 100 , alpha= 0.9);
# you can use grid() function to get grid lines
#plt.grid()
```

colors in matplotlib : b: blue g: green r: red c: cyan m: magenta y: yellow k: black w: white

## Histogram and Frequency Distribution


The histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

For example, let’s say that you have the following data about the age of 100 individuals:

```python
ages = ages = list(np.random.randint(low=18, high=80, size=100))

plt.figure(figsize=(8 , 6 ))  

# plotting the histogram 
plt.hist(ages , bins= 10 , edgecolor='black')
# bins=10: Divides the age data into 10 intervals (bins)
# edgecolor='black': Adds black borders around the bars to make them more distinct.


# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of 100 Individuals');
```

We can do the same thing with seaborn

```python
# Create a seaborn histogram
sns.histplot(ages, bins=10, kde=False)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of 100 Individuals')

# Show the plot
plt.show()
```

## Box Plot

It displays the distribution of the data based on quartiles and highlights outliers.

```python
sns.boxplot(x=ages)
plt.xlabel('Age')
plt.title('Box Plot of Ages')
plt.show()
```

## Heatmap

Displays data in a matrix format with color-coding to show values.

```python
# Create a sample correlation matrix
data = np.random.rand(10, 10)
sns.heatmap(data, annot=False)
plt.title('Heatmap Example')
plt.show()
```

## Bar Plot

Displays categorical data with bars.

```python
# Generate sample data
categories = ['A', 'B', 'C']
values = [10, 20, 30]

sns.barplot(x=categories, y=values)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Plot Example')
plt.show()
```
